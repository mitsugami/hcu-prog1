##集合・辞書に対する内包表記

リストだけでなく，集合・辞書も同様の内包表記が可能です．

【例題2-1】0から9までの整数の**集合**を作る

````python
# for文バージョン
y = set()   # y = {} ではダメ（辞書だと解釈されてしまう）
for x in range(10):
  y.add(x)  # y |= {x} と書いてもよい
print(y)
````

````python
# 内包表記バージョン
y = {x for x in range(10)}
print(y)
````

【例題2-2】0から9までの整数をキーとし，そのキーを文字列型にしたものを要素とする**辞書**を作る

````python
# for文バージョン
y = dict()  # y = {} でもOK
for x in range(10):
  y.update( {x: str(x)} )
print(y)
````

````python
# 内包表記バージョン
y = {x: str(x) for x in range(10)}
print(y)
````

##三項演算子

以下のようなif文は，
```
if B
  result = A
else
  result = C
```
以下のような式で表現することが可能です．
```
result = A if B else C
```
つまり，
```
(変数) = (条件がTrueのときの値) if (条件) else (条件がFalseのときの値)
```
という構文となり，これを「三項演算子」と言います．


###【例題4-1】
三項演算子を用いて，変数aが0以上なら positive ，負の値なら negative と表示するプログラムを作成しましょう．

````python
a = 10
s = 'positive' if a >= 0 else 'negative'
print(s)
````

###【例題4-2】
三項演算子をネストする（入れ子にする）ことで，複数の条件分けも可能です．

````python
a = 10
s = 'positive' if a > 0 else 'negative' if a < 0 else 'zero'
print(s)  # positive
````
